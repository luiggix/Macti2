# Transferencia de calor

Cuando estamos cocinando y por accidente tocamos una sart茅n caliente con la mano, el calor se transferir谩 a nuestra piel por contacto y puede causarnos una quemadura. Lo sabemos por experiencia, el calor en exceso puede da帽arnos. 

Pero 驴qu茅 es eso que llamamos calor?

El **calor** es una forma de energ铆a que se puede transferir de un sistema a otro como resultado de la diferencia en la temperatura. 

Es posible determinar las razones por las cuales esa forma de energ铆a se transfiere en diferentes medios; a estas razones o procesos se les denomina **transferencia de calor** y se pueden diferenciar tres tipos:

1. **Conducci贸n**: transferencia de calor basada en el contacto directo entre los cuerpos, sin intercambio de materia; el calor fluye desde un cuerpo de mayor temperatura a otro de menor temperatura que est谩 en contacto con el primero, se dice que hay un *gradiente de temperaturas*.
2. **Convecci贸n**: transferencia de calor de un lugar a otro debida al movimiento de fluidos. Este tipo de transferencia es com煤n en l铆quidos y gases, por ejemplo cuando se pone a hervir agua. 
3. **Radiaci贸n**: calor emitido en forma de ondas electromagn茅ticas, por ejemplo cuando se calienta un metal hasta llegar al rojo vivo.

## Modelo Matem谩tico

Ecuaci贸n de transferencia de calor:
$$
c_p \rho \frac{\partial T}{\partial t} +
c_p \rho \frac{\partial}{\partial x_j} \left( u_j T \right) -
\frac{\partial }{\partial x_j} \left( \kappa \frac{\partial T}{\partial x_j}\right) = 
S
$$

donde se define lo siguiente: $c_p$ es la capacidad calor铆fica espec铆fica, 
$\rho$ es la densidad, $\kappa$ es la conductividad t茅rmica, $S$ es la ganancia (fuente) o p茅rdida (sumidero) de calor y $\displaystyle \alpha = \frac{\kappa}{c_p \rho}$ es la difusividad t茅rmica.
Adem谩s, $x_j$ y $t$ representan las coordenadas cartesianas: $(x_1, x_2, x_3) \equiv (x, y, z)$ y el  tiempo, respectivamente; mientras que $u_j$ es la velocidad del medio y $T$ la temperatura. M谩s detalles de este modelo puede consultarlos en (Herrera, 2012).

Tomando en cuenta todas las hip贸tesis del modelo conceptual, el modelo matem谩tico a resolver es el siguiente:

$$
\begin{eqnarray}
-\kappa \frac{d^2 T}{d x^2} & = & S \qquad \text{Ecuaci贸n de conducci贸n de calor en 1D.}\\ \\
T(x = 0) & = & T_A  \qquad \text{Condiciones de}\\
T(x = L) & = & T_B \qquad \text{ de frontera}
\end{eqnarray}
$$

Obs茅rvese que se tienen condiciones de tipo *Dirichlet*: la variable dependiente, $T$, est谩 dada en las fronteras. Estas condiciones tambi茅n se conocen como de *primer tipo*.

## Modelo num茅rico

Aunque en este caso es posible obtener una soluci贸n anal铆tica, vamos a usar el m茅todo de diferencias finitas para aproximar una soluci贸n al modelo matem谩tico anterior.

### Discretizaci贸n del modelo matem谩tico.

Recordemos que el modelo matem谩tico consta de la siguiente ecuaci贸n:
$$
-\kappa \frac{d^2 T}{d x^2} = S \tag{1}
$$

Discretizamos la ecuaci贸n usando diferencias finitas de segundo orden:

- Consideramos un nodo $i$ de la malla, junto con sus vecinos $i+1$ e $i-1$, v茅ase la siguiente figura: 

<center>
<img src="./Figuras/ModCon05.png"  width='500px'/>
</center>

- Observe que todas las celdas son de la misma longitud $h$: la malla es *estructurada* y *uniforme*. 

La aproximaci贸n de la derivada en el nodo $i$ se escribe como sigue (Leveque, 2007):
$$
\left.\frac{d^2 T}{d x^2}\right|_i = \frac{T_{i+1} - 2 T_{i} + T_{i-1}}{h^2} + \mathcal{O}(h^2)
$$

Ahora sustituimos esta 煤ltima aproximaci贸n en la ecuaci贸n $(1)$ y rearreglamos t茅rminos para obtener:

$$
-r_i T_{i-1} + 2r_i T_{i} - r_i T_{i+1} = S_i \tag{2}
$$

donde $\displaystyle r_i = \frac{\kappa_i}{h^2}$ .

La ecuaci贸n (2) representa la conducci贸n de calor en el nodo $i$, y hace uso de sus vecinos $i+1$ e $i-1$. En esta ecuaci贸n, tanto $\kappa_i$ como $S_i$ representan la conductividad t茅rmica y la fuente en el nodo $i$.

En el caso que estamos estudiando, necesitamos calcular la temperatura en los nodos $i = 1, 2, 3, 4$, que son los nodos internos:

<center>
<img src="./Figuras/ModCon04.png"  width='500px'/>
</center>

Con base en la ecuaci贸n (2), debemos escribir una ecuaci贸n para cada uno de esos nodos:
$$
\begin{eqnarray*}
i = 1 \qquad \rightarrow \qquad -r_1 T_{0} + 2r_1 T_{1} - r_1 T_{2} & = & S_1 \\
i = 2 \qquad \rightarrow \qquad -r_2 T_{1} + 2r_2 T_{2} - r_2 T_{3} & = & S_2 \\
i = 3 \qquad \rightarrow \qquad -r_3 T_{2} + 2r_3 T_{3} - r_3 T_{4} & = & S_3 \\
i = 4 \qquad \rightarrow \qquad -r_4 T_{3} + 2r_4 T_{4} - r_4 T_{5} & = & S_4 
\end{eqnarray*}
$$

Este es un **sistema lineal** que se tiene que resolver para obtener la temperatura en cada uno de los nodos internos. Pero antes de hacerlo, debemos incluir las condiciones de frontera descritas en los modelos conceptual y matem谩tico.

Para $i=1$ tenemos la condici贸n $T_0  = T_A$:
- La ecuaci贸n para $i=1$ es: $-r_1 T_{0} + 2 r_1 T_{1} - r_1 T_{2} = S_1$
- Sustituyendo $T_0  = T_A$ obtenemos: $-r_1 \boxed{T_{A}} + 2 r_1 T_{1} - r_1 T_{2} = S_1$
- Pasando del lado derecho todo lo que es conocido obtenemos la ecuaci贸n para el nodo $i=1$ que incluye la condici贸n de frontera:

$$
2 r_1 T_{1} - r_{1} T_{2} = S_1 + r_{1} T_A \tag{3}
$$

Para $i=4$, donde tenemos la condici贸n $T_5 = T_B$,  seguimos el mismo procedimiento, de tal manera que la ecuaci贸n para este nodo, que incluye la condici贸n de frontera, es (verificarlo):

$$
-r_{4} T_{3} + 2 r_{4} T_{4} = S_4 + r_{4} T_B \tag{4}
$$

Usando las ecuaciones $(3)$ y $(4)$ obtenemos el sistema lineal final:

$$
\begin{eqnarray*}
2 r_1 T_{1} - r_{1} T_{2} & = & S_1 + r_{1} T_A \\
-r_2 T_{1} + 2r_2 T_{2} - r_2 T_{3} & = & S_2 \\
-r_3 T_{2} + 2r_3 T_{3} - r_3 T_{4} & = & S_3 \\
-r_{4} T_{3} + 2 r_{4} T_{4} & = & S_4 + r_{4} T_B 
\end{eqnarray*}
$$

En forma matricial se puede escribir el sistema lineal anterior como sigue:

$$
\underbrace{
	\left[
	\begin{matrix}
	2 & -1 & 0 & 0   \\
	-1 & 2 & -1 & 0  \\
	0 & -1 & 2 & -1  \\
	0 & 0 & -1 & 2    
	\end{matrix}
	\right]}_{{A}_{N \times N}}
\underbrace{
\left[
\begin{matrix}
T_1 \\ T_2 \\ T_3 \\ T_{4}
\end{matrix}
\right]}_{{T}_N} =
\underbrace{
\frac{1}{r}
\left[
\begin{matrix}
S_1 \\ S_2 \\ S_3 \\ S_{4}
\end{matrix}
\right] +
\left[
\begin{matrix}
T_A \\ 0 \\ 0 \\ T_B
\end{matrix}
\right]}_{{b}_N}
$$

Este sistema lineal es el que se tiene que resolver para obtener las temperaturas en los nodos $1,2,3$ y $4$.

## Modelo Computacional

Para construir un algoritmo que permita resolver el sistema lineal anterior, debemos ir paso a paso.

<p style="color:Navy">1. <strong>Escribir una funci贸n en Python que genere la matriz del sistema lineal.</strong> </p>

```python
# Se importan las bibliotecas Numpy y Matplotlib
import numpy as np
import matplotlib.pyplot as plt

#
#La siguiente funci贸n construye la matriz del sistema lineal
#
def buildMatrix(N):
    # Matriz de ceros
    A = np.zeros((N,N))

    # Primer rengl贸n
    A[0,0] = 2
    A[0,1] = -1
    # Renglones interiores
    for i in range(1,N-1):
        A[i,i] = 2
        A[i,i+1] = -1
        A[i,i-1] = -1
    # ltimo rengl贸n
    A[N-1,N-2] = -1
    A[N-1,N-1] = 2

    return A
```

```python
L  = 0.3   # Longitud del mango de la sart茅n
TA = 100.0 # Temperatura en el lado caliente
TB = 20.0  # Temperatura en el extremo del mango
  = 21.9  # Conductividad t茅rmica del titanio
S  = 0.0   # Fuente o sumidero

print("L = {}, TA = {}, TB = {},  = {}, S = {}".format(L, TA, TB, , S))
```

<p style="color:Navy">3. <strong>Definir los par谩metros num茅ricos.</strong> </p>

```python
N = 4          # N煤mero de nodos
h = L / (N+1)  # Espaciamiento entre los nodos
r =  / h**2   # Factor r

print("N = {}, h = {}, r = {}".format(N, h, r))
```

<p style="color:Navy">4. <strong>Definir los arreglos para almacenar la soluci贸n y para el lado derecho del sistema (RHS).</strong> </p>

```python
# Arreglo para almacenar la soluci贸n
T = np.zeros(N+2)
T[0]  = TA  # Frontera izquierda
T[-1] = TB  # Frontera derecha

# Lado derecho del sistema (RHS)
b = np.zeros(N)
b[:] = S / r    # Fuente o sumidero
b[0]  += T[0]   # Condici贸n de frontera izquierda
b[-1] += T[-1]  # Condici贸n de frontera derecha

print("T = {}".format(T))
print("b = {}".format(b))
```

<p style="color:Navy">5. <strong>Construir la matriz y resolver el sistema lineal.</strong> </p>

```python
# Construcci贸n de la matriz
A = buildMatrix(N)

# Soluci贸n del sistema lineal usando un m茅todo de Numpy.
T[1:N+1] = np.linalg.solve(A,b)
```

<p style="color:Navy">6. <strong>Imprimir y graficar la soluci贸n.</strong> </p>

```python
# Impresi贸n de la soluci贸n 
print(T)

# Graficaci贸n de la soluci贸n
x = np.linspace(0, L, N+2)
plt.figure(figsize=(8,4))
plt.plot(x, T, c='grey', lw=2.0)
plt.scatter(x, T, edgecolor='k', zorder= 10)
plt.xlabel('$x [m]$')
plt.ylabel('$T [^oC]$')
plt.grid()
plt.show()
```
